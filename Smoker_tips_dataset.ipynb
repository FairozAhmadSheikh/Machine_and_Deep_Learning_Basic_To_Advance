```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set the visual style for the plots
sns.set_theme(style="whitegrid")

# Load the dataset directly from seaborn's built-in repository
tips = sns.load_dataset("tips")
```

```python
# Display the first 5 rows and data info
print("Dataset Overview ")
print(tips.head())
print("\n Data Information ")
print(tips.info())
```

```python
# Summary statistics for numerical columns
print("\n--- Numerical Summary ---")
print(tips.describe())

# Check for missing values (there usually aren't any in this clean dataset)
print("\n--- Missing Values ---")
print(tips.isnull().sum())
# Value counts for categorical data to check for balance
print("\n--- Categorical Distribution ---")
print(f"Gender counts:\n{tips['sex'].value_counts()}")
print(f"Day of the week:\n{tips['day'].value_counts()}")
```

The Tip vs. Total Bill Relationship

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=tips, x="total_bill", y="tip", scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title("Relationship Between Total Bill and Tip Amount")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

Impact of Smoking Status and Gender

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True, palette="muted")
plt.title("Distribution of Total Bill by Day and Gender")
plt.show()
```

```python
# Creating a New Feature: Tip Percentage
tips['tip_pct'] = (tips['tip'] / tips['total_bill']) * 100

# Visualize Tip Percentage by Day
plt.figure(figsize=(10, 6))
sns.boxplot(data=tips, x="day", y="tip_pct", hue="smoker")
plt.title("Tip Percentage by Day: Smokers vs Non-Smokers")
plt.ylabel("Tip Percentage (%)")
plt.show()
```

```python
# Correlation Analysis
plt.figure(figsize=(8, 6))
numeric_tips = tips.select_dtypes(include=[np.number])
sns.heatmap(numeric_tips.corr(), annot=True, cmap="coolwarm", fmt=".2f")
# Use LaTeX for the title if you want to be fancy
plt.title(r"Correlation Heatmap of Numerical Features")
plt.show()
```
