### Method 3  recomended by matplotlib also known as Object Oriented Method

```python
fig,ax=plt.subplots()
ax.plot(x,y)
ax.set(title="Simple Line graph",
       xlabel='number',
       ylabel='square');
```

# Making plots from dictionaries

```python
car_prices={
    "creta":10,
    "scorpio":15,
    "breeza":16,
}
fig,ax=plt.subplots(figsize=(3,3))
ax.bar(car_prices.keys(),car_prices.values())
ax.set(title='Prices of cars',
       xlabel="Car Model",
       ylabel="Car Price")
ax.plot();
```

```python
# Horizontal bar graph requires data in list
fig,ax=plt.subplots(figsize=(3,3))
ax.barh(list(car_prices.keys()),list(car_prices.values()))
ax.set(title="Prices of cars and values",
       ylabel="carname",
       xlabel='prices ($)')
```
