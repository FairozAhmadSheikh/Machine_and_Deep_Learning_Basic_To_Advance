```python
import matplotlib.pyplot as plt 
```

```python
x=[1,2,3,4]
y=[v**2 for v in x ]
```

```python
plt.plot(x,y)
```

some methods for plotting 
# Method 1 

```python
fig=plt.figure()
ax=fig.add_subplot()
ax.plot(x,y)
plt.show()
```

Method 2 

```python
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show();
```

But we will be using `Method 3`

### Method 3  recomended by matplotlib also known as Object Oriented Method

```python
fig,ax=plt.subplots()
ax.plot(x,y)
ax.set(title="Simple Line graph",
       xlabel='number',
       ylabel='square');
```
