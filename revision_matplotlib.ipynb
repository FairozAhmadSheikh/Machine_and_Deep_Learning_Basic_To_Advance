# Making plots from dictionaries

```python
car_prices={
    "creta":10,
    "scorpio":15,
    "breeza":16,
}
fig,ax=plt.subplots(figsize=(3,3))
ax.bar(car_prices.keys(),car_prices.values())
ax.set(title='Prices of cars',
       xlabel="Car Model",
       ylabel="Car Price")
ax.plot();
```

```python
# Horizontal bar graph requires data in list
fig,ax=plt.subplots(figsize=(3,3))
ax.barh(list(car_prices.keys()),list(car_prices.values()))
ax.set(title="Prices of cars and values",
       ylabel="carname",
       xlabel='prices ($)')
```

#  Two Options for subplots

```python
# Method 1 : preferd method
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                       ncols=2,
                                       figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.randint(10),np.random.randint(10));
ax3.barh(list(car_prices.keys()),list(car_prices.values()));
ax4.hist(np.random.randn(1000));
```

```python
# Method 2 of plotting 
fix,ax=plt.subplots(nrows=2,
                     ncols=2,
                    figsize=(8,8))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(car_prices.keys(),car_prices.values());
ax[1,1].hist(np.random.randn(10000));
```

# Lets use a simple dataset as the graph shown for the bigger dataset will be more compilacted 

```python
car_data=pd.read_csv("car-sales.csv")
car_data.head()
```

```python
car_data['Price']=car_data['Price'].str.replace(r'[\$\,\.]',"",regex=True)
car_data.head()
```

```python
car_data['Price']=car_data['Price'].str[:-2]
car_data
```

```python
car_data['Sale_date']=pd.date_range("01/01/2025",periods=10)
```

```python
car_data
```

```python
car_data["Total_sales"]=car_data['Price'].astype(int).cumsum()
```

```python
car_data
```

```python
car_data.plot(x='Sale_date',y='Total_sales')
```
