```python
import matplotlib.pyplot as plt 
```

```python
x=[1,2,3,4]
y=[v**2 for v in x ]
```

```python
plt.plot(x,y)
```

some methods for plotting 
# Method 1 

```python
fig=plt.figure()
ax=fig.add_subplot()
ax.plot(x,y)
plt.show()
```

Method 2 

```python
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show();
```

But we will be using `Method 3`

### Method 3  recomended by matplotlib also known as Object Oriented Method

```python
fig,ax=plt.subplots()
ax.plot(x,y)
ax.set(title="Simple Line graph",
       xlabel='number',
       ylabel='square');
```

```python
# save plots
fig.savefig("saved_figure_line_plot_square_and_number.png")
```

```python
import numpy as np
x=np.linspace(0,10,40)
```

```python
fig,ax=plt.subplots()
ax.plot(x,x**2)  # Line PLot
ax.set(title="Numbers and ssquare",
       xlabel='Numbers',
       ylabel='Squared')
```

```python
# Scatter Plot 
x=np.linspace(0,100,10)
fig,ax=plt.subplots(figsize=(2,2))
ax.scatter(x,x**2)
```

```python
fig,ax=plt.subplots(figsize=(2,2))
ax.plot(x,np.cos(x))
```
